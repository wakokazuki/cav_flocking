# Settings

```python
data_label = "TOTAL" # 車両同士+障害物との衝突回数
speed = 4
num = 3
```

# Ave

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os
import sys

import heatmap_vel_agentnum

plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Arial'
# plt.rcParams['font.family'] = 'Meiryo'
```

```python
# print
list_of_files = glob.glob(f'../out/heatmap/csv/sp{speed:02}_num{num:02}/*agentnum_{data_label}.csv')
for ind, file in enumerate(list_of_files):
    print(file)
    df = pd.read_csv(file, index_col=0)
    if ind == 0:
        sum_count = df.values
    else:
        count = df.values
        sum_count += count
```

```python
sum_df = pd.DataFrame(sum_count, index=[3,4,5,6,7,8,9], columns=[4.0,6.0,8.0,10.0,12.0])
sum_df /= 10

sum_df 
```

```python
x_label = heatmap_vel_agentnum.x_param_label
y_label = heatmap_vel_agentnum.y_param_label
plt.figure(figsize=(9,6))
ax = sns.heatmap(sum_df, cmap='Reds', linewidths=2, annot=True, vmin=0, vmax=250, fmt=".1f") 
if data_label == "TOTAL":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{coll}}+F^{\mathrm{obs}}$")
elif data_label == "COLL":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{coll}}$")
elif data_label == "OBS":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{obs}}$")
ax.invert_yaxis()
plt.xlabel(x_label)
plt.ylabel(y_label)
# ax.collections[0].colorbar.set_label("衝突回数")
# plt.xlabel("最大速度 (m/s)")
# plt.ylabel("車両台数")
plt.title("average value")
plt.tight_layout()
plt.savefig(f'../out/heatmap/img/average_sp{int(speed):02}_num{int(num):02}_{data_label}.png', dpi=150)
plt.show()
```

```python
# Save as CSV
sum_df.to_csv(f"../out/heatmap/csv/ave_sp{int(speed):02}_num{int(num):02}_{data_label}.csv")
```

# Min

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os
import sys

import heatmap_vel_agentnum

plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Arial'
# plt.rcParams['font.family'] = 'Meiryo'
```

```python
# print
list_of_files = glob.glob(f'../out/heatmap/csv/sp{int(speed):02}_num{int(num):02}/*agentnum_{data_label}.csv')
for ind, file in enumerate(list_of_files):
    print(file)
    df = pd.read_csv(file, index_col=0)
    if ind == 0:
        min_count = df.values
    else:
        count = df.values
        for i in range(min_count.shape[0]):
            for j in range(min_count.shape[1]):
                if min_count[i,j] > count[i,j]:
                    min_count[i,j] = count[i,j]
```

```python
min_df = pd.DataFrame(min_count, index=[3,4,5,6,7,8,9], columns=[4.0,6.0,8.0,10.0,12.0])
min_df
```

```python
x_label = heatmap_vel_agentnum.x_param_label
y_label = heatmap_vel_agentnum.y_param_label
plt.figure(figsize=(9,6))
ax = sns.heatmap(min_df, cmap='Reds', linewidths=2, annot=True, vmin=0, vmax=250, fmt=".3g") 
if data_label == "TOTAL":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{coll}}+F^{\mathrm{obs}}$")
elif data_label == "COLL":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{coll}}$")
elif data_label == "OBS":
    ax.collections[0].colorbar.set_label(r"$F^{\mathrm{obs}}$")
ax.invert_yaxis()
plt.xlabel(x_label)
plt.ylabel(y_label)
# ax.collections[0].colorbar.set_label("衝突回数")
# plt.xlabel("最大速度 (m/s)")
# plt.ylabel("車両台数")
plt.title("minimum value")
plt.tight_layout()
plt.savefig(f'../out/heatmap/img/minimum_sp{int(speed):02}_num{int(num):02}_{data_label}.png', dpi=150)
plt.show()
```

```python
# Save as CSV
min_df.to_csv(f"../out/heatmap/csv/min_sp{int(speed):02}_num{int(num):02}_{data_label}.csv")
```
