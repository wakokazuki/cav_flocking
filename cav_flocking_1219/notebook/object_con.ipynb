```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import glob
import os
import sys

# sys.path.append(os.path.abspath(".."))


plt.rcParams['font.size'] = 16
plt.rcParams['font.family'] = 'Arial'
```

```python
# fileの一覧を表示
method_label = "nsgaiii"
pop_label = 120
gen_label = 300

# 条件を満たすファイルのリストを作成
files = glob.glob(f"../out/csvii/sumup/{method_label}_pop{pop_label}_gen{gen_label}_*.csv")
fies = files.sort()
for file in files:
    print(file)

#評価関数値の総和を追加
count = 0
OUTPUT = f'{method_label}_sum_loop{count}.csv'
for file in files:
    df = pd.read_csv(file, header=0,usecols=[12,13,14],dtype = {'CORR':'float','COLL':'float','OBS':'float'})
    df['Total'] = df.sum(axis=1,numeric_only=True)
    df = df.drop("CORR",axis = 1)
    df = df.drop("COLL",axis = 1)
    df = df.drop("OBS",axis = 1)
    df.to_csv(OUTPUT, index=True)
    count += 1
    OUTPUT = f'{method_label}_sum_loop{count}.csv'
print(df)

count = 0
Total_files = glob.glob(f"{method_label}_sum_*.csv")
for file in Total_files:
    if count == 0:
        total_df1 = pd.read_csv(file, header=0,usecols=[1],dtype = {"Total":"float"})
    else:
        df = pd.read_csv(file, header=0,usecols=[1],dtype = {"Total":"float"})
        total_df1 = pd.concat([total_df1, df], axis=1, ignore_index=True)
    count += 1
    
total_df1['mean'] = total_df1.mean(axis=1,numeric_only=True)
total_df1 = total_df1.drop(0,axis = 1)
total_df1 = total_df1.drop(1,axis = 1)
total_df1 = total_df1.drop(2,axis = 1)
total_df1 = total_df1.drop(3,axis = 1)
total_df1 = total_df1.drop(4,axis = 1)
total_df1 = total_df1.drop(5,axis = 1)
total_df1 = total_df1.drop(6,axis = 1)
total_df1 = total_df1.drop(7,axis = 1)
total_df1 = total_df1.drop(8,axis = 1)
total_df1 = total_df1.drop(9,axis = 1)
print(total_df1)
```
